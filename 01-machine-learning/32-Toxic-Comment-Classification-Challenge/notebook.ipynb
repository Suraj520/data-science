1. Naive Bayes (NB):<br> 
A probabilistic classifier based on the assumption that characteristics are conditionally independent given the class label. The posterior probability of each class is calculated using Bayes' theorem given the input information. The Naive Bayes formula is expressed as follows:<br><br>
<pre><code>
P(C_k|X) = (P(X|C_k) * P(C_k)) / P(X)
  </code></pre>
<br>
Where:<br>
- `P(C_k|X)` is the posterior probability of class `C_k` given the input `X`.<br>
- `P(X|C_k)` is the likelihood of observing the input `X` given class `C_k`.<br>
- `P(C_k)` is the prior probability of class `C_k`.<br>
- `P(X)` is the evidence probability, which acts as a normalizing factor.<br>
<br>
Naive Bayes assumes that the features `X` are conditionally independent given the class `C_k`. This allows us to calculate the likelihood term as the product of individual feature probabilities:
<br>
<pre><code>
P(X|C_k) = P(x_1|C_k) * P(x_2|C_k) * ... * P(x_n|C_k)
  </code></pre>
<br>
To classify a new instance, the classifier calculates the posterior probability for each class and selects the class with the highest probability as the predicted class.<br>
<br>
2. Multinomial Naive Bayes (MNB):<br>
MNB is created particularly for features that reflect discrete counts or frequencies, and is commonly used for text classification issues. The features are assumed to be produced by a multinomial distribution. The Multinomial Naive Bayes formula is similar to the Naive Bayes formula, but the likelihood term is calculated differently.<br>
In Multinomial Naive Bayes, the likelihood term is derived using the multinomial distribution, which indicates the probability of seeing a certain count for each feature given the class. The likelihood term is expressed by the following formula: <br><br>
<pre><code>
P(X|C_k) = (P(x_1|C_k)^n_1) * (P(x_2|C_k)^n_2) * ... * (P(x_n|C_k)^n_n)
  </code></pre>
<br>
Where:<br>
- `P(x_i|C_k)` is the probability of observing count `x_i` for feature `i` given class `C_k`.<br>
- `n_i` is the count of feature `i` in the input.<br>
<br>
The prior probability and posterior probability calculations in Multinomial Naive Bayes are the same as in Naive Bayes.<br>
<br>
3. Gaussian Naive Bayes (GNB):<br>
Gaussian Naive Bayes assumes that features follow a Gaussian (normal) distribution and estimates the mean and variance of each feature by class. The Gaussian naive Bayes formula can be expressed as: 
<br>
<pre><code>
P(C_k|X) = (P(X|C_k) * P(C_k)) / P(X)
  </code></pre>
<br>
Gaussian Naive Bayes probability terms are computed using the probability density function (PDF) of the Gaussian distribution. PDF computes the probability of observing a particular value for each feature in the class. The formula for the probability term can be expressed as:<br>
<pre><code>

  </code></pre>
<br>
Where:<br>
- `Î¼_ki` is the mean of feature `i` for class `C_k`.<br>
- `Ïƒ_ki` is the standard deviation of feature `i` for class `C_k`.<br>
- `x_i` is the value of feature `i` in the input.<br>
<br>
The computation of prior and posterior probabilities in Gaussian Naive Bayes is the same as in Naive Bayes. <br><br>
To brief, Naive Bayes assumes conditional independence of features, Multinomial Naive Bayes is suitable for features with discrete numbers or frequencies, and Gaussian Naive Bayes assumes features follow a Gaussian distribution. Each variant uses a different method for estimating the probability of observing a given feature in the class and can be classified based on the calculated probability.<br></div>

Importing modules
    </div>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
```

Loading the training dataset    </div>

```python
train_data = pd.read_csv('/kaggle/input/jigsaw-toxic-comment-classification-challenge/train.csv.zip')
```

ðŸ“Œ
Preprocessing the dataset    </div>

```python
comments = train_data['comment_text']
labels = train_data[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']]
```

<div style="background-color:#F0E3D2; color:#19180F; font-size:15px; font-family:Verdana; padding:10px; border: 2px solid #19180F; border-radius:10px"> 
ðŸ“Œ
Splitting the dataset into train and val sets    </div>

```python
train_comments, val_comments, train_labels, val_labels = train_test_split(comments, labels, test_size=0.2, random_state=42)
```

<div style="background-color:#F0E3D2; color:#19180F; font-size:15px; font-family:Verdana; padding:10px; border: 2px solid #19180F; border-radius:10px"> 
ðŸ“Œ
Defining the pipeline for multinomial naive bayes    </div>

```python
mnb_pipeline = Pipeline([
    ('vect', CountVectorizer()),
    ('clf', MultinomialNB())
])
```

<div style="background-color:#F0E3D2; color:#19180F; font-size:15px; font-family:Verdana; padding:10px; border: 2px solid #19180F; border-radius:10px"> 
ðŸ“Œ
Defining the hyperparameters to tune    </div>

```python
parameters = {
    'vect__ngram_range': [(1, 1), (1, 2)],  # Uni-gram or Bi-gram
    'clf__alpha': [0.1, 1.0, 10.0]  # Smoothing parameter
}
```

<div style="background-color:#F0E3D2; color:#19180F; font-size:15px; font-family:Verdana; padding:10px; border: 2px solid #19180F; border-radius:10px"> 
ðŸ“Œ
Performing grid search for hyperparameter tuning and displaying individual classification report    </div>

```python
mnb_grid_search = GridSearchCV(mnb_pipeline, parameters, cv=5, n_jobs=-1)
for label in labels.columns:
    mnb_grid_search.fit(train_comments, train_labels[label])
    best_mnb_model = mnb_grid_search.best_estimator_
    val_predictions = best_mnb_model.predict(val_comments)
    print(f"Label: {label}")
    print(classification_report(val_labels[label], val_predictions))
```

<div style="background-color:#F0E3D2; color:#19180F; font-size:15px; font-family:Verdana; padding:10px; border: 2px solid #19180F; border-radius:10px"> 
ðŸ“Œ
Generating submission file    </div>

```python
submission_data = pd.read_csv('/kaggle/input/jigsaw-toxic-comment-classification-challenge/test.csv.zip')
submission_comments = submission_data['comment_text']
```

```python
from tqdm import tqdm
submission_df = pd.DataFrame({'id': submission_data['id']})
for label in tqdm(labels.columns):
    best_mnb_model.fit(comments, labels[label])
    label_submission = best_mnb_model.predict_proba(submission_comments)
    submission_df[label] = label_submission[:, 1]

submission_df.to_csv('submission.csv', index=False)
```
